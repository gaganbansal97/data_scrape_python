```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.makaan.com/price-trends/property-rates-for-rent-in-bangalore"
response = requests.request("GET", url)

# parse the downloaded data
data = BeautifulSoup(response.text, 'html.parser')
price_data = data.find_all('tr', attrs= {'itemtype': 'http://schema.org/Place'})
# total number of cards
for c in price_data:  
    loc = c.find('a', attrs= {'class': 'link-block'})
    price = c.find_all('td', attrs= {'class': 'caps ta-c'})
    print(loc.text, "-", price[0].text,"-",price[1].text,"-",price[2].text)
```
